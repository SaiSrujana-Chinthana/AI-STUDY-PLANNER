```python
# ===============================
# AI STUDY PLANNER GENERATOR
# ===============================

import pandas as pd
from datetime import datetime, timedelta

def generate_study_plan(subjects, weak_subjects, exam_date, hours_per_day=6):

    today = datetime.today().date()
    exam_date = datetime.strptime(exam_date, "%Y-%m-%d").date()

    days_left = (exam_date - today).days

    if days_left <= 0:
        print("‚ùå Exam date must be in the future!")
        return None

    subject_list = [s.strip() for s in subjects.split(",")]
    weak_list = [s.strip() for s in weak_subjects.split(",")]

    plan = []

    for day in range(days_left):
        current_date = today + timedelta(days=day)
        remaining_hours = hours_per_day

        daily_schedule = {}

        for subject in subject_list:
            if subject in weak_list:
                hours = 2
            else:
                hours = 1

            daily_schedule[subject] = hours
            remaining_hours -= hours

        plan.append({
            "Date": current_date,
            **daily_schedule
        })

    df = pd.DataFrame(plan)
    return df


# ===============================
# USER INPUT
# ===============================

subjects = input("Enter subjects (comma separated): ")
weak_subjects = input("Enter weak subjects (comma separated): ")
exam_date = input("Enter exam date (YYYY-MM-DD): ")
hours_per_day = int(input("Enter total study hours per day: "))


# ===============================
# GENERATE PLAN
# ===============================

study_plan = generate_study_plan(subjects, weak_subjects, exam_date, hours_per_day)

if study_plan is not None:
    print("\n‚úÖ Study Plan Generated Successfully!\n")
    display(study_plan)
```

```python
study_plan.to_csv("study_plan.csv", index=False)
print("üìÅ File saved as study_plan.csv")
```
